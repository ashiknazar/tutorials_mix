# enumerate

enumerate() allows us to iterate through a sequence but it keeps track of both the index and the element. The enumerate() function takes in an iterable as an argument, such as a list, string, tuple, or dictionary

```python
l1=['red','blue','green']
s1="yellow"

obj1=enumerate(l1)
obj2=enumerate(s1)

print(list(obj1))
print(list(obj2))

print(list(enumerate(s1,3)))
```

```python
from abc import ABC, abstractmethod

# Abstract class with abstract methods
class Animal(ABC):
    #@abstractmethod
    def make_sound(self):
        pass

# Concrete classes inheriting from the abstract class
class Dog(Animal):
    def make_sound(self):
        return "Woof!"

class Cat(Animal):
    def make_sound(self):
        return "Meow!"

# Function using the abstract class without knowing the concrete implementation
def animal_sound(animal):
    return animal.make_sound()

# Usage example
dog = Dog()
cat = Cat()

print("Dog says:", animal_sound(dog))
print("Cat says:", animal_sound(cat))
```
