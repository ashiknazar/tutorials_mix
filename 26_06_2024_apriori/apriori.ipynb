```python
def inspect(results):
    lhs=[tuple(result[2][0][0])[0] for result in results]#item1
    rhs=[tuple(result[2][0][1])[0] for result in results]#item2
    
    support=[result[1] for result in results]
    confidence=[result[2][0][2] for result in results]
    lift=[result[2][0][3] for result in results]
    
    return list(zip(lhs,rhs,support,confidence,lift))
```

```python
import pandas as pd
rules=apriori(transactions=transactions,min_support=0.003,min_confidence=0.01,min_lift=3,min_length=2,max_length=2)
result=list(rules)
result
def create_dataframe(data):
    items = []
    support = []
    base_items = []
    add_items = []
    confidence = []
    lift = []
    for record in data:
        for ordered_stat in record.ordered_statistics:
            items.append(record.items)
            support.append(record.support)
            base_items.append(ordered_stat.items_base)
            add_items.append(ordered_stat.items_add)
            confidence.append(ordered_stat.confidence)
            lift.append(ordered_stat.lift)
    df = pd.DataFrame({
        'items': items,
        'support': support,
        'base_items': base_items,
        'add_items': add_items,
        'confidence': confidence,
        'lift': lift
    })

    return df
''' 
from collections import namedtuple
RelationRecord = namedtuple('RelationRecord', ['items', 'support', 'ordered_statistics'])
OrderedStatistic = namedtuple('OrderedStatistic', ['items_base', 'items_add', 'confidence', 'lift'])
'''

df = create_dataframe(result)
```
