```python
x=data.drop(['Gender','User ID','Purchased'],axis=1)
x.values
```

```python
y=data['Purchased']
y.values
```

```python
#we can also using via iloc
#x=data,iloc[:,2,4].values
#y=data.iloc[:,4].values
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=4)
```

```python
x_test
```

```python
x_test_dataf=pd.DataFrame(x_test)
x_test_dataf
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
#predicting test set

y_pred=lr.predict(x_test)
y_pred
```

A confusion matrix is a tabular summary of the number of correct and incorrect predictions made by a classifier.



```python
#Confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
#visualize the confusion matrix with seaborn heatmap

cm_mat=pd.DataFrame(data=cm,columns=['Actual Positive:1,','Actual Negative:0'],
                   index=['Predict Positive:1','Predict Negative:0'])
sns.heatmap(data=cm_mat,annot=True,cmap='cividis_r')
```

```python
#confusin matrix evaluate for accuracy

from sklearn import metrics

print(metrics.accuracy_score(y_test,y_pred))
```

```python
#check the overfitting and underfitting

print(lr.score(x_train,y_train))
```

```python
print(lr.score(x_test,y_test))
```

```python
#classification report

from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```
