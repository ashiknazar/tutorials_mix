```python
import pandas as pd
```

```python
data=pd.read_csv('50_Startups.csv')
```

```python
data.head()
```

```python
data['State'].nunique()
```

```python
data.isnull().sum()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oneh=OneHotEncoder()
```

```python
onee=oneh.fit_transform(data['State'])
```

```python

oneh = OneHotEncoder()

state_reshaped = data['State'].values.reshape(-1, 1)

onee = oneh.fit_transform(state_reshaped)

onee_df = pd.DataFrame(onee.toarray(), columns=oneh.get_feature_names_out(['State']))

df = pd.concat([data, onee_df], axis=1)
```

```python
df.columns
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scal=StandardScaler()
```

```python
dd=scal.fit_transform(df[['R&D Spend', 'Administration', 'Marketing Spend','Profit']])
```

```python
dd
```

```python
scaled_df = pd.DataFrame(dd, columns=['R&D Spend_', 'Administration_', 'Marketing Spend_','Profit_'])
df = pd.concat([df, scaled_df], axis=1)
```

```python
df
```

```python
XX=df.iloc[:,[5,6,7,8,9,10]].values
```

```python
YY=df.iloc[:,11].values
```

```python
XX.shape
```

```python

from sklearn.model_selection import train_test_split 
x_tr,x_tes,y_tr,y_tes=train_test_split(XX,YY,test_size=0.2,random_state=4)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor(n_estimators=100,random_state=42)
```

```python
model.fit(x_tr,y_tr)
```

```python
y_pr=model.predict(x_tes)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_tes,y_pr)
```
