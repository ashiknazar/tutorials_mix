```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix
```

```python
data=pd.read_csv('Social_Network_Ads.csv')
```

```python
data.head()
```

```python
scl=StandardScaler()
XX=scl.fit_transform(data[['Age','EstimatedSalary']])
```

```python
YY=data.iloc[:,4].values
```

```python
x_train,x_test,y_train,y_test=train_test_split(XX,YY,test_size=0.3,random_state=4)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
confusion_matrix(y_test,y_pred)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression

# Assuming x_train and y_train are defined as:
# x_train: shape (280, 2)
# y_train: shape (280,)

# Step 1: Fit the logistic regression model
model = LogisticRegression()
model.fit(x_train, y_train)

# Step 2: Create a grid of values to evaluate the model
x_min, x_max = x_train[:, 0].min() - 1, x_train[:, 0].max() + 1
y_min, y_max = x_train[:, 1].min() - 1, x_train[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))

# Step 3: Compute the probabilities for the grid values
Z = model.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]
Z = Z.reshape(xx.shape)

# Step 4: Plot the data and the decision boundary
plt.figure(figsize=(10, 6))
plt.contourf(xx, yy, Z, alpha=0.8, cmap=plt.cm.Spectral)
plt.scatter(x_train[:, 0], x_train[:, 1], c=y_train, edgecolors='k', marker='o', cmap=plt.cm.Spectral)

# Add labels and title
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('Logistic Regression Decision Boundary')

# Show the plot
plt.colorbar()
plt.show()
```
