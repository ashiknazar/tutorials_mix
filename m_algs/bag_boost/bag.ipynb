```python
import pandas as pd
```

```python
data=pd.read_csv('hearts.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
X=data.iloc[:,:-1].values
```

```python
X
```

```python
Y=data.iloc[:,13].values
```

```python
Y
```

```python
from sklearn.model_selection import train_test_split
x_tr,x_tes,y_tr,y_tes=train_test_split(X,Y,test_size=0.2,random_state=43)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
```

```python
from sklearn.svm import SVC
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
from sklearn.metrics import accuracy_score
```

```python
log=LogisticRegression()
rnd_frst=RandomForestClassifier()
svm_cls=SVC()
naiv=GaussianNB()
dec_tree=DecisionTreeClassifier()
```

```python
for model in (log,rnd_frst,svm_cls,naiv,dec_tree):
    model.fit(x_tr,y_tr)
    y_pred=model.predict(x_tes)
    print(model.__class__.__name__,' ',end='')
    print(accuracy_score(y_tes,y_pred))
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bag=BaggingClassifier(GaussianNB(),n_estimators=100,max_samples=100,bootstrap=True,n_jobs=1,oob_score=True,max_features=0.7)
bag.fit(x_tr,y_tr)
y_pr=bag.predict(x_tes)
print(accuracy_score(y_tes,y_pr))
```

```python
for model in (log,rnd_frst,svm_cls,naiv,dec_tree):
    bag=BaggingClassifier(model,n_estimators=100,max_samples=100,bootstrap=True,n_jobs=1,oob_score=True,max_features=0.7)
    bag.fit(x_tr,y_tr)
    y_pr=bag.predict(x_tes)
    print(bag.base_estimator_,accuracy_score(y_tes,y_pr))
```
