```python
import numpy as np
import pandas as pd

# Generate random data for demonstration
np.random.seed(0)
n_samples = 1000

# Generate random weights (in kg) and heights (in cm)
weights = np.random.normal(loc=70, scale=10, size=n_samples)
heights = np.random.normal(loc=170, scale=10, size=n_samples)

# Create a pandas DataFrame
data = pd.DataFrame({'Weight': weights, 'Height': heights})

# Display the first few rows of the dataset
print(data.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize the distribution of weight and height
sns.boxplot(data=data)
plt.xlabel('Features')
plt.ylabel('Values')
plt.title('Boxplot of Weight and Height')
plt.show()
```
