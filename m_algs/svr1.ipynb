```python
def onehot(data,column_list,imputer='ignore'):
    from sklearn.preprocessing import OneHotEncoder
    from sklearn.impute import SimpleImputer
    import numpy as np
    import pandas as pd
    for column_name in column_list:
        imput=SimpleImputer(missing_values=np.NaN,strategy='most_frequent')
        data[column_name]=imput.fit_transform(data[column_name].values.reshape(-1,1))
        
        column = data[column_name].values.reshape(-1, 1)
        
        # Initialize OneHotEncoder
        encoder = OneHotEncoder()

        # Fit and transform the  column
        column_encoded = encoder.fit_transform(column)

        # Convert the sparse matrix to a dense array and assign column names
        encoded_df = pd.DataFrame(column_encoded.toarray(), columns=encoder.get_feature_names_out([column_name]))

        # Concatenate the encoded_df with the original DataFrame
        df_encoded = pd.concat([data, encoded_df], axis=1)

        # Drop the original 'sex' column
        df_encoded.drop(column_name, axis=1, inplace=True)
        data=df_encoded
    return data
        
```

```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder

def onehoti(data, column_list):
    for column_name in column_list:
        # Initialize and fit the SimpleImputer
        imput = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
        data[column_name] = imput.fit_transform(data[column_name].values.reshape(-1, 1))

        # Initialize and fit the OneHotEncoder
        encoder = OneHotEncoder(sparse=False)
        column_encoded = encoder.fit_transform(data[column_name].values.reshape(-1, 1))

        # Create a DataFrame with encoded data and appropriate column names
        encoded_df = pd.DataFrame(column_encoded, columns=encoder.get_feature_names_out([column_name]))

        # Concatenate the original DataFrame with the encoded DataFrame
        data = pd.concat([data.drop(column_name, axis=1), encoded_df], axis=1)

    return data

# Examp
```
