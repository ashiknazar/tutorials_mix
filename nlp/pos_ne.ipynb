```python
import nltk
import pandas as pd

words = set()

for para in dat:
    wor = nltk.word_tokenize(para)
    for x in wor:
        words.add(x)


pos_tags = nltk.pos_tag(list(words))


chunks = nltk.ne_chunk(pos_tags, binary=False)


ents = []
lbls = []

for chunk in chunks:
    if hasattr(chunk, 'label'):
        ents.append(' '.join(c[0] for c in chunk))
        lbls.append(chunk.label())


ent_lbls = list(set(zip(ents, lbls)))
ent_df = pd.DataFrame(ent_lbls, columns=['ents', 'lbls'])

print(ent_df)
```
